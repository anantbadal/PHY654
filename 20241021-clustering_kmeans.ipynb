```python
import matplotlib.pyplot as plt

x = [1, 2, 4, 4, 5, 5, 10, 4, 3, 11, 14 , 6, 10, 12, 14, 13, 11, 12, 12, 12]
y = [17, 17, 18, 21, 18, 19, 24, 17, 16, 25, 24, 22, 21, 21, 22, 24, 21, 24, 21, 22]

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans
data = list(zip(x, y))
```

```python
print (data)
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
